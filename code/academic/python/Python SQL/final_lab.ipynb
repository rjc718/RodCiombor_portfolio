<b><u>Exercise 17.1 e</u></b>
<br>
<b>Insert a new title for an author. Remember that the book must have an entry in the author_ISBN table and an entry in the titles table.</b>

```python
new_isbn = '0131234567'
author_id = 6
titles_values = (new_isbn, 'The Art of Coding', 1, '2026')
author_isbn_values = (author_id, new_isbn)

cursor = connection.cursor()
cursor = cursor.execute(
    'INSERT INTO titles (isbn, title, edition, copyright) VALUES (?, ?, ?, ?)',
    titles_values
)
cursor = cursor.execute(
    'INSERT INTO author_ISBN (id, isbn) VALUES (?, ?)',
    author_isbn_values
)

pd.read_sql('SELECT t.title, t.edition, t.copyright, t.isbn FROM titles t INNER JOIN author_ISBN ai ON t.isbn = ai.isbn WHERE ai.id =? ORDER BY t.title', 
    connection,
    params = [author_id],
    index_col=['title']
)
```
